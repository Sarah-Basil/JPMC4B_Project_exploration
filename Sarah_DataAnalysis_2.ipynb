```python
# Visualizing missing values:

plt.figure(figsize=(12,8))  # trying to get a large enough fig size

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
# ^ heatmap to visualize any missing values

# inverting y-axis for better readability
plt.gca().invert_yaxis()

# rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.title('Missing Data Heatmap')
plt.show()
```

```python
print("shape: ", df.shape, '\n') # to see what the general shape of the data is

# now to check for the missing values:
nan_count = np.sum(df.isnull(), axis = 0)

condition = nan_count != 0 # looking for all columns with missing values

col_names = nan_count[condition].index # index gets the column names
nan_cols = list(col_names) # converting column names to list to view them better
nan_col_types = df[nan_cols].dtypes # in order to also print the datatypes of the missing vals
print("missing value columns: ")
for col, dtype in nan_col_types.items():  # trying to iterate using items()
  print(f"{col}: {dtype}")
print("\n")

nan_count # just to look at the overall count
```

```python
target_count = df['TARGET'].value_counts()
print(target_count)

# Visualizing the distribution
sns.countplot(df['TARGET'])
plt.title('Distribution of Target Variable')
plt.show()
```
