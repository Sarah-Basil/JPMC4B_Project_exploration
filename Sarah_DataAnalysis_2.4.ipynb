```python
print("shape: ", df.shape, '\n') # to see what the general shape of the data is

# now to check for the missing values:
nan_count = np.sum(df.isnull(), axis = 0)
missing_col_num = nan_count[nan_count != 0].count();
total_nan_values = nan_count.sum()
print("Number of missing features columns: ", missing_col_num)
print("Number of missing values overall: ", total_nan_values, "\n")

condition = nan_count != 0 # looking for all columns with missing values

col_names = nan_count[condition].index # index gets the column names
nan_cols = list(col_names) # converting column names to list to view them better
nan_col_types = df[nan_cols].dtypes # in order to also print the datatypes of the missing vals
print("missing value columns: ")
for col, dtype in nan_col_types.items():  # trying to iterate using items()
  print(f"{col}: {dtype}")
print("\n")

nan_count[nan_count != 0] # just to look at the overall count
```

```python
target_count = df['TARGET'].value_counts()
print(target_count)

# # Visualizing the distribution
# sns.countplot(df['TARGET'])
# plt.title('Distribution of Target Variable')
# plt.show() <-- hard to do, because too many features
```

```python
# Correlation matrix for numerical columns (just to get an initial idea)
numerical_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numerical_cols].corr()

# getting the correlation of all numerical features with 'TARGET' (what we want to predict)
target_corr = corr_matrix['TARGET'].sort_values(ascending=False)

# displaying each feature with its correlation value:
for feature, value in target_corr.items():
    print(f"Feature: {feature},\nCorrelation with TARGET: {value:.4f}\n\n")
```

```python
# Plotting the correlations with 'TARGET'
plt.figure(figsize=(12, 20))  # Increase the figure size for better readability
sns.barplot(x=target_corr.values, y=target_corr.index, palette='coolwarm_r', hue=None, legend=False)

# adjusting the font size and title for readability (there are so many featuers)
plt.title('Initial Correlation of Numerical Features with TARGET', fontsize=16)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Features', fontsize=12)

# Adding some padding to the plot to avoid crowded labels
plt.tight_layout()

plt.show()
```

```python
# Find percentage of missing values per column
missing_percentage = (df.isnull().sum() / df.shape[0]) * 100

# Display columns with more than 40% missing values and their percentage
columns_to_drop = missing_percentage[missing_percentage > 40]
print("Columns to drop due to high missing percentage and their missing values percentage:")
for col, perc in columns_to_drop.items():
    print(f"{col}: {perc:.2f}%")
```

```python
df = df.drop(columns = columns_to_drop) # dropping the columns that have > 40% missing data
```

```python
print(df.shape)
df.head()
```

I would like to utilize some of these techniques... like MICE or using models to predict values (like KNN model to impute values for example).. but the dataset is so large that this just seems computationally expensive and a bit unfeasible. We can see if it's possible later, but for now maybe I must stick to standard simpler imputation procedures like filling with mean, median, mode

```python
# For the rest, we can try to impute missing values for remaining features
# For numerical columns, we could potentially use the median
numerical_cols = df.select_dtypes(include=[np.number]).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# For categorical columns, we can maybe use the mode (or most frequent value)
categorical_cols = df.select_dtypes(include=[object]).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# ^ these were simpler imputing methods, but maybe I'll try a slightly more complex one to see how it is.
```
