```python
# Find percentage of missing values per column
missing_percentage = (df.isnull().sum() / df.shape[0]) * 100

# Display columns with more than 40% missing values and their percentage
columns_to_drop = missing_percentage[missing_percentage > 40]
columns_to_drop_index = missing_percentage[missing_percentage > 40].index  # Corrected: Use index to get column names

print("Columns to drop due to high missing percentage and their missing values percentage:")
for col, perc in columns_to_drop.items():
    print(f"{col}: {perc:.2f}%")
```

```python
# Impute EXT_SOURCE_2 with mean
df['EXT_SOURCE_2'].fillna(df['EXT_SOURCE_2'].mean(), inplace=True)

# Impute EXT_SOURCE_3 using Iterative Imputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

impute_features = ['EXT_SOURCE_3', 'EXT_SOURCE_2', 'DAYS_BIRTH', 'AMT_INCOME_TOTAL', 'AMT_CREDIT']

imputer = IterativeImputer(random_state=0)
df[impute_features] = imputer.fit_transform(df[impute_features])
```

```python
from sklearn.impute import KNNImputer #importing KNNImputer to use on important features

# since there are a small amt. of values to impute, I will try to handle them individually:

# imputing 'AMT_ANNUITY' with median:
df['AMT_ANNUITY'].fillna(df['AMT_ANNUITY'].median(), inplace=True)

# imputing 'AMT_GOODS_PRICE' with median:
df['AMT_GOODS_PRICE'].fillna(df['AMT_GOODS_PRICE'].median(), inplace=True)

# imputing 'NAME_TYPE_SUITE' with mode:
df['NAME_TYPE_SUITE'].fillna(df['NAME_TYPE_SUITE'].mode()[0], inplace=True)
# ^indexing to [0] in order to take the first mode that is found

# imputing 'OCCUPATION_TYPE' by creating a new category 'Unknown':
df['OCCUPATION_TYPE'].fillna('Unknown', inplace=True)

# imputing 'CNT_FAM_MEMBERS' with mode (since you would typically consider most frequent)
df['CNT_FAM_MEMBERS'].fillna(df['CNT_FAM_MEMBERS'].mode()[0], inplace=True)

# trying to impute 'EXT_SOURCE_2' and 'EXT_SOURCE_3' using KNN Imputer:

# # initializing KNN Imputer:
# imputer = KNNImputer(n_neighbors=5)

# # fitting and transforming the data to impute 'EXT_SOURCE_2' and 'EXT_SOURCE_3:
# df[['EXT_SOURCE_2', 'EXT_SOURCE_3']] = imputer.fit_transform(df[['EXT_SOURCE_2', 'EXT_SOURCE_3']])

# imputing social circle variables with median:
social_circle_cols = ['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE',
                      'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']
for col in social_circle_cols:
    df[col].fillna(df[col].median(), inplace=True)

# imputing 'DAYS_LAST_PHONE_CHANGE' with median:
df['DAYS_LAST_PHONE_CHANGE'].fillna(df['DAYS_LAST_PHONE_CHANGE'].median(), inplace=True)

# imputing AMT_REQ_CREDIT_BUREAU_*' variables with median
amt_req_cols = ['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY',
                'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_MON',
                'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR']

for col in amt_req_cols:
    df[col].fillna(df[col].median(), inplace=True)

# 10. (Optional) Create missing indicators if you think missingness is informative
# For example, create indicator for missing 'OCCUPATION_TYPE'
# Since we have filled missing 'OCCUPATION_TYPE' with 'Unknown', this step may not be necessary.
# If you decide to create an indicator before filling missing values:
# df['OCCUPATION_TYPE_missing'] = df['OCCUPATION_TYPE'].isnull().astype(int)

# After imputation, check if any missing values remain
print("Remaining missing values per column after imputation:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
df = df[df['CODE_GENDER'] != 'XNA']
df['CODE_GENDER'].replace('XNA', df['CODE_GENDER'].mode()[0], inplace=True) #preparing CODE_GENDER for binary mapping
```

```python
# encoding binary variables using binary mapping
binary_mapping = {
    'NAME_CONTRACT_TYPE': {'Cash loans': 0, 'Revolving loans': 1},
    'FLAG_OWN_CAR': {'N': 0, 'Y': 1},
    'FLAG_OWN_REALTY': {'N': 0, 'Y': 1},
    'CODE_GENDER': {'F': 0, 'M': 1},
}

for col, mapping in binary_mapping.items():
    df[col] = df[col].map(mapping)
```

```python
# ordinal encoding for education feature
education_mapping = {
    'Lower secondary': 1,
    'Secondary / secondary special': 2,
    'Incomplete higher': 3,
    'Higher education': 4,
    'Academic degree': 5,
    'Missing': 0  # Assign 0 to 'Missing'
}

df['NAME_EDUCATION_TYPE'] = df['NAME_EDUCATION_TYPE'].map(education_mapping)
```

```python
# For ORGANIZATION_TYPE
top_10_orgs = df['ORGANIZATION_TYPE'].value_counts().nlargest(10).index
df['ORGANIZATION_TYPE'] = df['ORGANIZATION_TYPE'].apply(lambda x: x if x in top_10_orgs else 'Other')
df = pd.get_dummies(df, columns=['ORGANIZATION_TYPE'], drop_first=True)

# For OCCUPATION_TYPE
top_10_occs = df['OCCUPATION_TYPE'].value_counts().nlargest(10).index
df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].apply(lambda x: x if x in top_10_occs else 'Other')
df = pd.get_dummies(df, columns=['OCCUPATION_TYPE'], drop_first=True)


#handling high cardinal data separately, might also do frequency encoding
```
